```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("data_intern.csv")
```

```python
data.head(5)
```

```python
data.replace({"diagnosis":{"M":1,"B":2}},inplace=True)
```

```python
data.head(5)
```

```python
data.drop("Unnamed: 32",axis=1,inplace=True)
```

```python
data.dropna(inplace=True)
```

```python
data.head(5)
```

```python
data.shape
```

```python
data.drop("id",axis=1,inplace=True)
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=data.drop("diagnosis",axis=1)
```

```python
x.shape
```

```python
y=data["diagnosis"]
```

```python
y=y.values
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder=OneHotEncoder(sparse=False)
```

```python
y=encoder.fit_transform(y.reshape(-1,1))
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y)
```

```python
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
model=Sequential()
```

```python
model.add(Dense(30,input_shape=(30,),activation="relu"))
model.add(Dense(20,activation="relu"))
model.add(Dropout(0.8))
model.add(Dense(15,activation="relu"))
model.add(Dense(7,activation="relu"))
model.add(Dense(2,activation="softmax"))
```

```python
earlystop=EarlyStopping(verbose=1,patience=25,mode="min",monitor="val_loss")
```

```python
model.compile(optimizer="adam",loss="binary_crossentropy")
```

```python
model.fit(x=x_train,y=y_train,validation_data=(x_test,y_test),epochs=30,callbacks=earlystop)
```

```python
loss_df=pd.DataFrame(model.history.history)
```

```python
loss_df.plot(kind="line")
```

```python
def detect(x,y):
    x_train,x_test,y_train,y_test=train_test_split(x,y)
    model.fit(x=x_train,y=y_train,validation_data=(x_test,y_test),callbacks=earlystop,epochs=30)
    loss_df=pd.DataFrame(model.history.history)
    loss_df.plot(kind="line")
```

```python
detect(x,y)
```
