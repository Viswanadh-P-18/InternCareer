# Exploratory data analysis

```python
import seaborn as sns
```

```python
c=data.corr()
```

```python
sns.heatmap(c,annot=True)
```

```python
sns.pairplot(data.drop(["Links","Country","Rank","Username"],axis=1))
```

TOP 5 Channels based on likes

```python
data.groupby(["Username"])["Likes"].mean().sort_values(ascending=False)[0:5].plot(kind="bar")
```

# TOP 5 CHANNELS :SUBSCRIBERS

```python
data.groupby(["Username"])["Suscribers"].mean().sort_values(ascending=False)[0:5].plot(kind="bar")
```

# TOP channels based on visits

```python
data.groupby(["Username"])["Visits"].mean().sort_values(ascending=False)[0:5].plot(kind="bar")
```

```python
data.columns
```

# EDA based on categories

```python
data.groupby(["Categories"])["Visits"].mean().sort_values(ascending=False)[0:5].plot(kind="bar")
```

```python
data.groupby("Country")["Visits"].mean().sort_values(ascending=False)[0:5].plot(kind="bar")
```
